```python
# 使うライブラリ
# !pip install japanize-matplotlib

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()
```

## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_hist_data = pd.read_csv("data/hist_data.csv")
plt.hist(df_hist_data["aデータ"], bins = 30, color = "y", alpha = 0.5, label = "aデータ")   #alpha...透明度
plt.hist(df_hist_data["bデータ"], bins = 30, color = "c", alpha = 0.5, label = "bデータ")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=1)  #アンカーを右上にして、凡例枠の右上をアンカーに合わせる。余白をあける。
plt.title("aデータとbデータの温度に関する度数分布")
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_scatter_data = pd.read_csv("data/scatter_data.csv")

class0 = df_scatter_data[df_scatter_data["class"] == 0]
class1 = df_scatter_data[df_scatter_data["class"] == 1]
class2 = df_scatter_data[df_scatter_data["class"] == 2]

plt.scatter(class0['length (cm)'], class0['width (cm)'], color = "r", label = "class0")
plt.scatter(class1['length (cm)'], class1['width (cm)'], color = "b", label = "class1")
plt.scatter(class2['length (cm)'], class2['width (cm)'], color = "g", label= "class2")

plt.xlabel('length (cm)')
plt.ylabel('width (cm)')

plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=1) 
plt.grid(True)  #グリッド線の追加

plt.title('クラス０〜２のlengthとwidthに関する散布図')
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_line_data = pd.read_csv("data/line_data.csv")

plt.plot(df_line_data["期間"], df_line_data["件数"], marker = "o")  #markerの形を指定できる。全部で３５種類ある。

plt.xlabel("期間")
plt.ylabel("件数")

plt.grid(True)
plt.xticks(rotation = 30)  #x軸のラベルを３０°回転させる

plt.title("期間ごとの件数の推移")
```

```python
plt.bar(df_line_data["期間"], df_line_data["費用"])

plt.xlabel("期間")
plt.ylabel("費用（円）")

plt.grid(True)
plt.xticks(rotation = 30) 

plt.title("期間ごとの費用の推移")
```
